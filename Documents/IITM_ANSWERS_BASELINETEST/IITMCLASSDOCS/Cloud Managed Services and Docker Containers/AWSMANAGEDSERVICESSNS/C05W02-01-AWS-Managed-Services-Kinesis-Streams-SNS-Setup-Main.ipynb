# <font color='blue'> SNS Setup </font>

AWS provides several managed services. The Simple Notification Service (SNS) is one such service that provides a mechanism for event notifications to be delivered to users and other services. In our implementation, we will make use of SNS to provide email notifications. Let us now set up the mechanism.

## <font color='blue'> Topic Creation </font>

Use your credentials to login into your AWS educate account. On the AWS management console page, search for ```SNS```, and select ```Simple Notification Service```. 

SNS-1.JPG

On this page, select the ```Topics``` tab available in the section to the left:

SNS-2.JPG


When you click on ```Topics```, it will display the available/created topics using SNS services. Click on ```Create Topic```. 

On the next page select the ```Standard``` option to create the topic. 

Enter the name for your topic.

Enter the display name. 

Do not change any other default options, and click on ```Create```. 

Successful creation will lead to a page similar to this:

SNS-3.JPG

## <font color='blue'> Creating A Subscription </font>

Click on ```Create Subscription``` on the same page, which leads to this: 

SNS-4.JPG


Click on `Protocol`, and select the `Email` option. In the `Endpoint` field, enter your email ID, which is where SNS will send event notifications.

Do not change any other default options, and click on `Create Subscription`:

SNS-5.JPG

A successful subscription will take you to a page like this:

SNS-6.JPG


You will receive a confirmation mail at the email ID you provided. Within that email, click on the ```Confirm Submission``` link: 

SNS-7.JPG

Once subscription is confirmed, a fresh page pops up: 

SNS-8.JPG

## <font color='blue'> Publish, And Notification </font>

Now, go to the newly created topic and click on ```Publish Message```. On the following page, fill in the ```Subject``` and ```Message Body``` fields. Do no change any other default content, and click on ```Publish```: 

SNS-9.JPG

You should receive a message at the email ID you provided:

SNS-10.JPG

We have demonstrated how the AWS Simple Notification Service can be set up, and deliver event notifications to you!
