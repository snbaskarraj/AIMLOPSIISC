# <font color='blue'> Uber Case Study </font>

Apache Spark is a cornerstond of the Big Data infrastructure at Uber.

Uber currently runs more than one hundred thousand Spark applications per day, across multiple different compute environments.

## <font color='blue'> Uber Maps Data Collection </font>

The Maps Data Collection team at Uber captures images of street signs to improve the efficiency and quality of our maps data in order to facilitate:

* A more seamless trip experience for both riders and driver-partners
* Taking pictures of food items to be displayed within the Uber Eats app

The resulting imagery and metadata is processed via a series of Apache Spark jobs, and the results are stored in a Hadoop Distributed File System (HDFS) instance.

The images needed to be efficiently accessed by downstream Apache Spark jobs, as well as randomly accessed by operators responsible for map edits.

M02W07-Notebook-Uber-Maps-Data-Architecture.png

Different file formats for each of ingestion, intermediate, and final output data.

During ingestion, the contents of the files are unaltered so that additional processing and transformations may be done in parallel via Spark, with little or no data filtering.

Therefore, a simple file format is used that provides optimal write performance and does not have the overhead of schema-centric file formats.

Intermediate data generated by a previous task is most often read in its entirety and is filtered by neither columns nor rows.

Due to the lower write overhead, Avro files are generally used to store intermediate data when:

* The data is written once and read once.
* The reader does not filter the data.
* The reader reads the data sequentially.

The final output could be one of 

* Imagery metadata: Read multiple times, and filtered - Parquet format
* Imagery: Binary data with intensive resource processing - Avro format
* Aggregated data: Combined information about both metadata nd imagery - JSON format

## <font color='blue'> Uber Spark Compute Service (uSCS) </font>

The ever-growing infrastructure at Uber means environments are constantly changing.

This makes it quite difficult for users to get reliable access to data sources, compute resources, and supporting tools. 

Also, it becomes more challenging for the data team to communicate these environmental changes to users.

Problems with using Apache Spark at scale include:

* Data source diversity: Spark applications access multiple data sources - HDFS, Apache Hive, Apache Cassandra, and MySQL - with diverse and ever-changing confogurations
* Multiple compute clusters: Uberâs compute platform provides support for Spark applications across multiple types of clusters, both in on-premises data centers and the cloud
* Multiple Spark versions: Some versions of Spark have bugs, donât work with particular services, or have yet to be tested on  Uber's compute platform
* Dependency issues: As the number of applications grow, so too does the number of required language libraries deployed to executors

The Uber Spark Compute Service (uSCS) is a Spark-as-a-service solution to help manage the complexities of running Spark at this scale.

Uber maintains compute infrastructure across regions, each with its own copy of important storage services - such as HDFS - and several compute clusters.

M02W07-Notebook-Uber-uSCS-Uber-Infrastructure.png

Because storage is shared within a region, an application that runs on one compute cluster should run on all other compute clusters within the same region.

However, differences in resource manager functionality mean that some applications will not automatically work across all compute cluster types

One of the goals of uSCS is to enable Spark to work seamlessly over our entire large-scale, distributed data infrastructure by abstracting these differences away.

To use uSCS, a user or service submits an HTTP request describing an application to the Gateway, which intelligently decides where and how to run it, then forwards the modified request to Apache Livy.

M02W07-NOtebook-Uber-uSCS-Architecture.png

The uSCS Gateway makes rule-based decisions to modify the application launch requests it receives, and tracks the outcomes that Apache Livy reports.

Multiple Apache Livy deployments run per region at Uber, each tightly coupled to a particular compute cluster.

Apache Livy submits each application to a cluster and monitors its status to completion.

Advantages of this architecture include:

* The uSCS abstracts configuration information for diverse data sources
* uSCS enables tracking every application on the compute platform
* Migrating applications running old Spark versions to newer versions - is now automated

# <font color='blue'> Informatica: IoT Data Ingestion And Predictive Analytics </font>

Streaming and IoT data management includes the capabilities to ingest, manage, and act on a variety of real-time data.
 
Apache Spark brings massive scale capabilities and the recently introduced Structured Streaming to help solve real world streaming analytics use cases.

There are two key streaming use case patterns:

* Load the data onto a data lake for data science and advanced analytics
* Run pre-created predictive models on streaming data through a pipe in real time, for fraud detection

Informatica offers end-to-end streaming data management, starting from ingestion of streaming and IoT data to applying reasoning on the data, to finally acting on the streaming data.

M02W07-Notebook-Informatics-Streaming-IoT-Analytics-Architecture.png

Informatica offers both cloud-ready and cloud-native streaming and IoT data ingestion solutions from streaming sources like Kafka, IoT, web logs, etc. 

Data is ingested into 

* Cloud data lakes such as Amazon S3 and Azure Data Lake Storage (ADLS), or 
* Messaging systems such as Kafka and Amazon Kinesis

Informatica Big Data Management (BDM) helps customers perform transformations and aggregations on the data stored in the lake, using the highly scalable Apache Spark engine.

Informatica Bog Data Streaming (BDS) uses the power of Spark Streaming engine for performing analytics on streaming data at very high volumes.

Big Data Streaming adopted Spark Structured Streaming to allow feature such as event time windowing, and the ability to handle late arrival of events.
