# <font color='blue'> AWS Machine To Cloud Connectivity Framework </font>

The Machine to Cloud Connectivity Framework solution provides secure factory equipment connectivity to the AWS Cloud that reduces downtime.

The features of this framework include:

* Fast and robust data ingestion
* Highly reliable and durable storage of equipment data
* Serverless event-driven applications that help manage a factory's configuration.

The advantages it offers include:

* Ease of deployment
* Help driving an increase in factory efficiency and uptime
* Improving production flexibility, and identify new business opportunities.

The Machine to Cloud Connectivity Framework solution uses **AWS IoT** services to collect data from connected devices and send that data to the cloud. 

One can automatically deploy the following architecture using this solution, and the accompanying AWS CloudFormation template:

image.png

The AWS CloudFormation template solution deloys the following components:

* **AWS IoT Core** topics, **AWS IoT Greengrass**, and **AWS Lambda** functions to manage communication between factory equipment and the AWS Cloud. 
* **Amazon Kinesis Data Firehose** and **Amazon Simple Storage Service (Amazon S3)** to store the data from the AWS IoT Core topics
* **Amazon DynamoDB** and **Amazon S3** to store job files and job metadata.

A typical flow of information in this solution architecture is as follows:

1. The user creates a job request that contains 
    * Information about the equipment
    * The data they want to read from the equipment, and
    * The frequency at which they want to read the data
2. Lambda: 
    * Translates the request into machine-readable format
    * Pushes the translated request to the connected device, and 
    * Stores the request on the device gateway running AWS IoT Greengrass
3. Lambda running in AWS IoT Greengrass: 
    * Reads the translated request
    * Establishes a connection between the device gateway and the equipment
    * Gets the data defined in the request, and 
    * Sends it to a topic in AWS IoT Core
    
Once the data is translated to human-readable format and posted to a topic in AWS IoT Core, one can specify an action to perform on the data using the Rules Engine.

# <font color='blue'> AWS Connected Vehicle Solution </font>

Amazon Web Services (AWS) enables automotive manufacturers and suppliers to build serverless IoT applications that manage connected vehicle data, without having to manage any infrastructure.

With AWS IoT, customers can connect vehicles and devices to the AWS Cloud securely, with low latency and with low overhead.

The solution offers connectivity to the AWS Cloud, and a framework that helps customers integrate AWS IoT and AWS Greengrass into the Automotive Grade Linux (AGL) software stack.

You can build upon this framework to address a variety of use cases: 

* Voice interaction, navigation and other location-based services
* Remote vehicle diagnostics and health monitoring
* Predictive analytics and required maintenance alerts
* Vehicle safety and security services

The AWS Connected Vehicle Solution architecture is organized as shown:

image.png

When AWS IoT receives a message:

1. It authenticates and authorizes the message 
2. The Rules Engine executes the appropriate rule on the message
3. The Rule Engine then routes the message to the appropriate backend application.

Here is the run-time flow of execution within the AWS Connected vehicle Solution:

1. An AWS IoT rule sends telematics data to an **Amazon Kinesis Data Firehose** delivery stream, which encrypts and streams raw vehicle telematics data to an **Amazon S3 bucket**.
2. If an **Amazon Kinesis Data Analytics** application detects an anomaly, the record is sent to **Amazon Kinesis Data Streams**.
3. The **Amazon Kinesis Data Streams** component: 
    * Invokes an **AWS Lambda** function that parses the record
    * Stores it in an **Amazon DynamoDB** table, and 
    * Triggers an **Amazon Simple Notification Service (Amazon SNS)** notification to users.

# <font color='blue'> AWS Connected Mobility Solution </font>

The AWS Connected Mobility Solution (CMS) is a reference architecture that aids automotive OEMs and suppliers in their journey to becoming mobility service providers.

AWS CMS includes:

* Software modules to securely manage the lifecycle of in-vehicle edge compute
* Machine learning inference models
* cloud data ingest
* Storage
* Business logic, and 
* APIs to power applications in the cloud.

Production deployments include:

* e-Scooter fleets with tens of thousands of units
* Global rental car and OEM fleets scaling into the millions of connected vehicles

image.png


The Benefits that AWS CMS offers include:

* Quickly build connected vehicle systems using well architected services, with scaling, from scratch
* Gain knowledge from AWS’ experience working with several automotive suppliers and manufacturers
* Reduce cost per connected vehicle by moving from dedicated on-prem IT infrastructure, to serverless cloud resources

# <font color='blue'> uReplicator: Uber Engineering’s Robust Apache Kafka Replicator </font>

Uber use Apache Kafka as a message bus to collect system and application logs, as well as event data from the rider and driver apps. 

This data is then made available to a variety of downstream consumers via Kafka.

image.png


Given the large-scale use of Kafka, Uber ends up using multiple clusters in different data centers, often needing a global view of the data:
    * To compute business metrics related to trips, they need to gather information from all data centers and analyze it in one place.

Historically, Uber has used the open source MirrorMaker tool shipped with the Kafka package, to replicate data across data centers:


image.png


Mirror Maker uses a high-level Kafka consumer to fetch the data from the source cluster, which it feeds into a Kafka producer to dump it into the destination cluster.

This leads to many scalability issues:

* Expensive rebalancing: Violates Uber's end-to-end latency guarantee
* Difficulty adding topics: requires a static whitelist of topics that can be added, which also leads to expensive rebalancing
* Possible data loss: Automatic offset commits are unreliable
* Metadata sync issues

Ultimately, Uber developed a solution named **uReplicator** to solve these issues:

image.png

uReplicator’s various components work in different ways toward reliability and stability.

1. The **Helix uReplicator Controller** uses **Zookeeper** to fulfil all its responsibilities:
    * Distribute and assign topic partitions to each worker process
    * Handle addition/deletion of topics/partitions
    * Handle addition/deletion of uReplicator workers
    * Detect node failures and redistribute those specific topic-partitions 
2.  A **uReplicator Worker** replicates topic partitions from source cluster to destination cluster, and avoids rebalancing
3.  A **Helix Agent** for each uReplicator notifies the **Dynamic Kafka Consumer** of changes, to add/remove topic-partitions
4. A **Dynamic Kafka Consumer** instance on each uReplicator worker, which 
    * Removes the rebalance part, and 
    * Adds a mechanism to add/delete topic-partitions on the fly
    
The benefits of uReplicator include:

* **Stability**: Rebalancing now happens only during startup and node addition/deletion, and only affects a few partitions
* **Easier Scalability**: Adding a new node to an existing cluster is now much simpler
* **Easier Operation**: No need to restart the cluster when adding/deleting/expanding Kafka topics
* **Zero Data Loss**: uReplicator commits checkpoints only after the data has been persisted on the destination cluster
