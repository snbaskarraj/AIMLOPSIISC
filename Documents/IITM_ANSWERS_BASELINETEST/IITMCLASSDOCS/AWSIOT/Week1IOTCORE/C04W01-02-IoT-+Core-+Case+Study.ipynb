# <font color='blue'>  Smart Product Solution </font>

This is a generic architecture for cloud setup to support varied smart product solutions. These can be smart lighting products in the home or smart fitness and healthcare wearables, just to mention a few examples.

## <font color='blue'> Capabilities </font>


- Local computing within products
- Data and diagnostic information push to the cloud
- Telemetry information and visualization
- Web console for user dashboard and visualization
- Continuous security and safety monitoring based on periodic security rules based checks, and immediate alerts based on that
- Sophisticated event rule processing to trigger actions either at the system level, or back to the users
- Notifications to the users based on alerts or general events 
- Data storage for long-term analysis
